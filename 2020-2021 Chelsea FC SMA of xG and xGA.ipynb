```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/sagnikm/Documents/Python/2020-2021 xG and xGA Graphs/Full 38 games of 2020-2021/End of Season CFC.csv")
```

```python
df
```

```python
df.head()
```

```python
df["match"] = range(1,39)
```

```python
df["xgSMA"] = df["xG"]. rolling(window=3).mean()
df["xgaSMA"] = df["xGA"]. rolling(window=3).mean()
```

```python
plt.plot(df.match, df.xgSMA, color="blue")
plt.plot(df.match, df.xgaSMA, color="red")
```

```python
df["difference"] = df["xG"] - df["xGA"]
```

```python
plt.plot(df.match, df.difference, color="black")
```

```python
import numpy as np
```

```python
fig, ax = plt.subplots()
plt.plot(df.match, df.xgSMA, color="blue")
plt.plot(df.match, df.xgaSMA, color="red")
ax.set_title("Simple Moving Averages of xG and xGA")
ax.set_ylabel("Expected Goals")
ax.set_xlabel("Matchweek")
```

```python
fig, ax = plt.subplots()
plt.plot(df.match, df.difference, color="black")
ax.set_title("Simple Moving Averages of Difference Between xG and xGA")
ax.set_ylabel("Expected Goals")
ax.set_xlabel("Matchweek")
```
